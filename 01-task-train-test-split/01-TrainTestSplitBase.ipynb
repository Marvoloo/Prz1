```python
# осуществляем парсинг данных текстового файла как csv, хедеры "вытаскиваем" из текстового файла в массив
import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit
target_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data"

spam = pd.read_csv(target_url, header=None)
target_url1 = "https://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.names"
headers = pd.read_csv(target_url1, sep="\r", header=None)
headers = headers[30:]
headers = headers[0].str.split(":", expand=True)
headers = headers[0].to_numpy()
headers = np.append(headers, "class")
spam.columns = headers
print(spam)
```

```python
spam['class'].value_counts()
# Класс неполностью сбалансирован, т.к. значений класса 0 больше чем 1. 
# Но при разбиении на выборки обязательно нужно пользоваться "StratifiedShuffleSplit", чтобы сохранить пропорции классов в выборках
```

```python
# Разбиваем на выборки... 80/20
sss = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
y = spam["class"]
X = spam.drop("class", axis=1)
sss.get_n_splits(X, y)
sss.split(X,y)
for train_index, test_index in sss.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
print(X_train)
print(X_test)
print(y_train)
print(y_test)
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

```python
# Разбиваем на выборки... 70/30
sss = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=0)
y = spam["class"]
X = spam.drop("class", axis=1)
sss.get_n_splits(X, y)
sss.split(X,y)
for train_index, test_index in sss.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
print(X_train)
print(X_test)
print(y_train)
print(y_test)
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```
